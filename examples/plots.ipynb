```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder = "mnist"
# Load the data

mnist_noise = pd.read_csv(f"{folder}/noise.csv")
mnist_scales = pd.read_csv(f"{folder}/scales.csv")
mnist_test_noise = pd.read_csv(f"{folder}/test_noise.csv")

folder = "cifar10"
# Load the data
cifar_noise = pd.read_csv(f"{folder}/noise.csv")
cifar_scales = pd.read_csv(f"{folder}/scales.csv")
cifar_test_noise = pd.read_csv(f"{folder}/test_noise.csv")
```

```python
mnist_noise.columns = ["accuracy", "supervised loss", "jacobian loss", "total loss", "noise"]
mnist_scales.columns = ["accuracy", "supervised loss", "jacobian loss", "total loss", "lambda_JR", "noise", "fc1", "fc2"]
mnist_test_noise.columns = ["accuracy", "supervised loss", "jacobian loss", "total loss", 'lambda_JR', "dataset", "n_proj", "epochs", "batch_size", "learning_rate", "noise", "test_noise"]
```

```python
cifar_noise.columns = ["accuracy", "supervised loss", "jacobian loss", "total loss", "noise", "lambda_JR", "fc1", "fc2"]
cifar_test_noise.columns = ["accuracy", "supervised loss", "jacobian loss", "total loss", 'lambda_JR', "dataset", "n_proj", "epochs", "batch_size", "learning_rate", "noise", "test_noise"]
cifar_scales.columns = ["accuracy", "supervised loss", "jacobian loss", "total loss", "lambda_JR", "noise"]
```

```python
plt.plot(mnist_noise["noise"], mnist_noise["total loss"], label="Training Noise", color="blue")
plt.plot(mnist_scales['lambda_JR'], mnist_scales["total loss"], label="Regularization Parameter", color="green")
plt.plot(mnist_test_noise["test_noise"], mnist_test_noise["total loss"], label="Testing Noise", color="red")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.title("Loss vs Noise for MNIST")
plt.xlabel("Scale of Additive Noise")
plt.ylabel("Total Loss");
plt.savefig(f"{folder}/mnist_loss_noise.png")
```

```python
plt.plot(cifar_noise["noise"], cifar_noise["total loss"], label="Training Noise", color="blue")
plt.plot(cifar_scales['lambda_JR'], cifar_scales["total loss"], label="Regularization Parameter", color="green")
plt.plot(cifar_test_noise["test_noise"], cifar_test_noise["total loss"], label="Test Noise", color="red")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Scale of Additive Noise")
plt.title("Loss vs Noise for CIFAR-10")
plt.legend()
plt.ylabel("Total Loss");
plt.savefig(f"{folder}/cifar10_loss.png")
```

```python
plt.plot(cifar_noise["noise"], cifar_noise["supervised loss"], label="Training Noise", color="blue")
plt.plot(cifar_scales['lambda_JR'], cifar_scales["supervised loss"], label="Regularization Parameter", color="green")
plt.plot(cifar_test_noise["test_noise"], cifar_test_noise["supervised loss"], label="Test Noise", color="red")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Scale of Additive Noise")
plt.title("Supervised Loss vs Noise for CIFAR-10")
plt.legend()
plt.ylabel("Total Loss");
plt.savefig(f"{folder}/cifar10_supervised_loss.png")
```

```python
plt.plot(cifar_noise["noise"], cifar_noise["jacobian loss"], label="Training Noise", color="blue")
plt.plot(cifar_scales['lambda_JR'], cifar_scales["jacobian loss"], label="Regularization Parameter", color="green")
plt.plot(cifar_test_noise["test_noise"], cifar_test_noise["jacobian loss"], label="Test Noise", color="red")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Scale of Additive Noise")
plt.title("Jacobian Loss vs Noise for CIFAR-10")
plt.legend()
plt.ylabel("Jacobian Loss");
plt.savefig(f"{folder}/cifar10_jacobian_loss.png")
```
